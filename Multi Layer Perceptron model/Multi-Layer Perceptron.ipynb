```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('wine.data', sep = ',', header = None)
```

```python
data.head()
```

```python
col = ['class', 'alcohol', 'malic_acid', 'ash', 'alcalinity_of_ash', 'magnesium',
       'total_phenols', 'flavanoids', 'nonflavanoid_phenols', 'proanthocyanins',
       'color_intensity', 'hue', 'od280/od315_of_diluted wines', 'proline']
```

```python
wine = pd.DataFrame(data)
```

```python
wine.columns = col
```

```python
wine.head()
```

```python
wine.info()
```

```python
wine.shape
```

```python
wine.describe()
```

```python
x = wine.drop(['class'], axis = 1)
```

```python
y = wine['class']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (10, 5))

sns.heatmap(wine.corr(), annot = True, cmap = 'coolwarm').set_title('Wine Data Correlation')

plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

col_list = []

for i in wine.columns:
    if i != 'class':
        col_list.append(i)
        
z = wine[col_list]
vif_data = pd.DataFrame()
vif_data['features'] = z.columns
vif_data['VIF'] = [variance_inflation_factor(z.values, i) for i in 
                  range(len(z.columns))]

vif_data
```

```python
r_columns = vif_data[(vif_data.VIF >= 60)]
r_columns
```

```python
d_columns = [i for i in r_columns.features]
```

```python
d_columns
```

```python
x = x.drop(d_columns, axis = 1)
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_res = LogisticRegression()
```

```python
log_res.fit(x_train, y_train)
```

```python
logres_pred = log_res.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

print('Logistic Regression accuracy score:', round(accuracy_score(y_test, logres_pred), 2))

cmatrix = confusion_matrix(y_test, logres_pred)

sns.heatmap(cmatrix, annot = True, cmap ='coolwarm',
            xticklabels = [1,2,3], yticklabels = [1,2,3]).set_title('Logistic Regression Confusion Matrix')

plt.show()
```

#### Random Forest Regression

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 20)
```

```python
rfc.fit(x_train, y_train)
```

```python
rfc_pred = rfc.predict(x_test)
```

```python
print('Random Forest Accuracy Score', round(accuracy_score(y_test, rfc_pred), 2))

cmatrix2 = confusion_matrix(y_test, rfc_pred)

sns.heatmap(cmatrix2, annot = True, cmap = 'coolwarm',
            xticklabels = [1,2,3], yticklabels = [1,2,3]).set_title('Random Forest Confusion Matrix')

plt.show()
```

```python
n_input = 20

shape = x_train.shape[1]

n_hidden_1 = 15

n_hidden_2 = 15

n_hidden_3 = 15

classes = 3
```

```python
import tensorflow as tf
from keras.layers import Dense, Dropout
from keras.models import Sequential
from tensorflow.keras.utils import to_categorical
```

```python
import livelossplot
plot_losses = livelossplot.PlotLossesKeras()
```

```python
model = Sequential()

model.add(Dense(n_input, activation = 'relu', name = 'Input_Layer', input_shape = (shape, )))

model.add(Dense(n_hidden_1, activation = 'relu', name = 'Hidden_layer_1'))

model.add(Dropout(0.25))

model.add(Dense(n_hidden_2, activation = 'relu', name = 'Hidden_layer_2'))

model.add(Dropout(0.25))

model.add(Dense(n_hidden_3, activation = 'relu', name = 'Hidden_layer_3'))

model.add(Dense(classes, activation = 'softmax', name = 'Output_layer'))
```

```python
model.summary()
```

```python
# Compile model
model.compile(loss = 'categorical_crossentropy',
              optimizer = 'adam',
              metrics = ['accuracy'])
```

```python
y_train_cat = y_train - 1
y_test_cat = y_test - 1
```

```python
y_train_cat = to_categorical(y_train_cat, 3)
y_test_cat = to_categorical(y_test_cat, 3)
```

```python
# Train the model
model.fit(x_train, y_train_cat,
         batch_size = 5,
         epochs = 30,
         callbacks = [plot_losses],
         verbose = 0,
         validation_data = (x_test, y_test_cat))
```

```python
score = model.evaluate(x_test, y_test_cat, verbose = 0)
print('Test Loss:', score[0])
print('Test accuracy:', score[1])
```
