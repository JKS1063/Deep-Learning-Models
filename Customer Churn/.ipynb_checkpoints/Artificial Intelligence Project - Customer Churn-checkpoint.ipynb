## C. Model Building

```python
import livelossplot
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import confusion_matrix

from keras.layers import Dense, Dropout
from keras.models import Sequential
import tensorflow as tf
```

```python
le = LabelEncoder()
df.Churn = le.fit_transform(df.Churn)
```

```python
sc = StandardScaler()
```

```python
plot_loss = livelossplot.PlotLossesKeras()
```

#### a. Build a sequential model using Keras, to find out if the customerwouldchurn or not, usingâ€˜tenureâ€™ as the feature and â€˜Churnâ€™ as the dependent/target column:
* i. The visible/input layer should have 12 nodes with â€˜Reluâ€™ as activation function.
* ii. This model would have 1 hidden layer with 8 nodes and â€˜Reluâ€™ as activation function
* iii. Use â€˜Adamâ€™ as the optimization algorithm
* iv. Fit the model on the train set, with number of epochs to be 150
* v. Predict the values on the test set and build a confusion matrix
* vi. Plot the â€˜Accuracy vs Epochsâ€™ graph

```python
x1 = df[['tenure']]
y1 = df['Churn']
```

```python
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, test_size = 0.2, shuffle = True, random_state = 42)
```

```python
x1_train = sc.fit_transform(x1_train)
x1_test = sc.fit_transform(x1_test)
```

```python
shape1 = x1_train.shape[1]
```

```python
model1 = Sequential()

model1.add(Dense(12, activation = 'relu', name = 'Input_Layer', input_shape = (shape1, )))
model1.add(Dense(8, activation = 'relu', name = 'Hidden_Layer'))
model1.add(Dense(1, activation = 'sigmoid', name = 'Output_Layer'))

model1.summary()
```

```python
model1.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
x1_train.shape, y1_train.shape
```

```python
history1 = model1.fit(x1_train, y1_train,
          epochs = 150,
          callbacks = [plot_loss],
          verbose = 0,
          validation_data = (x1_test, y1_test))
```

```python
score = model1.evaluate(x1_test, y1_test, verbose = 0)
print('Test Loss:', round(score[0], 2))
print('Test Accuracy:', round(score[1], 2))
```

```python
y1_pred = model1.predict(x1_test)
```

```python
y1_pred
```

```python
y1_pred = np.round(y1_pred) 
```

```python
cm = confusion_matrix(y1_test, y1_pred)
sns.heatmap(cm, annot = True, cmap = 'coolwarm', fmt = 'g')
```

```python
epochs = 150
plt.plot(history1.history['accuracy'], label = 'Accuracy')
plt.plot(history1.history['val_accuracy'], label = 'Validation_Accuracy')
plt.xticks(np.arange(0, epochs, 10))
plt.legend(loc = 'upper right')
plt.show()
```

#### b. Build the 2nd model using same target and feature variables:
* i. Add a drop-out layer after the input layer with drop-out value of 0.3
* ii. Add a drop-out layer after the hidden layer with drop-out value of 0.2
* iii. Predict the values on the test set and build a confusion matrix
* iv. Plot the â€˜Accuracy vs Epochsâ€™ graph

```python
x2 = df[['tenure']]
y2 = df['Churn']

x2_train, x2_test, y2_train, y2_test = train_test_split(x2 ,y2, test_size = 0.3, shuffle = True, random_state = 42)

x2_train = sc.fit_transform(x2_train)
x2_test = sc.fit_transform(x2_test)

shape2 = x2_train.shape[1]
```

```python
model2 = Sequential()

model2.add(Dense(12, activation = 'relu', name = 'Input_Layer', input_shape = (shape2,)))
model2.add(Dropout(0.3))
model2.add(Dense(8, activation = 'relu', name = 'Hidden_Layer'))
model2.add(Dropout(0.2))
model2.add(Dense(1, activation = 'sigmoid', name = 'Output_Layer'))

model2.summary()
```

```python
model2.compile(loss = 'binary_crossentropy',
               optimizer = 'adam',
               metrics = ['accuracy'])
```

```python
history2 = model2.fit(x2_train, y2_train,
           epochs = 150,
           callbacks = [plot_loss],
           verbose = 0,
           validation_data = (x2_test, y2_test))
```

```python
score2 = model2.evaluate(x2_test, y2_test)
print('Test Loss:', round(score2[0], 2))
print('Test Accuracy:', round(score2[1], 2))
```

```python
y2_pred = model2.predict(x2_test)
y2_pred = np.round(y2_pred)
y2_pred
```

```python
cm2 = confusion_matrix(y2_test, y2_pred)
sns.heatmap(cm2, annot = True, cmap = 'coolwarm', fmt = 'g')
plt.show()
```

```python
plt.plot(history2.history['accuracy'], label = 'Accuracy')
plt.plot(history2.history['val_accuracy'], label = 'Validation Accuracy')
plt.xticks(np.arange(0, epochs, 10))
plt.legend(loc = 'upper right')
plt.show()
```

#### c. Build the 3rd model using â€˜Tenureâ€™, â€™Monthly Chargesâ€™ & â€˜Total Chargesâ€™ as the features and â€˜Churnâ€™ as the dependent/target column:
* i. The visible/input layer should have 12 nodes with â€˜Reluâ€™ as activation function.
* ii. This model would have 1 hidden layer with 8 nodes and â€˜Reluâ€™ as activation function
* iii. Use â€˜Adamâ€™ as the optimization algorithm
* iv. Fit the model on the train set, with number of epochs to be 150
* v. Predict the values on the test set and build a confusion matrix

```python
df.columns
```

```python
x3 = df[['tenure', 'MonthlyCharges', 'TotalCharges']]
```

```python
y3 = df.Churn
```

```python
x3_train, x3_test, y3_train, y3_test = train_test_split(x3, y3, test_size = 0.3, shuffle = True, random_state = 42)
```

```python
x3_train = sc.fit_transform(x3_train)
x3_test = sc.fit_transform(x3_test)
```

```python
shape3 = x3_train.shape[1]
```

```python
model3 = Sequential()

model3.add(Dense(12, activation = 'relu', name = 'Input_Layer', input_shape = (shape3, )))
model3.add(Dense(8, activation = 'relu', name = 'Hidden_Layer'))
model3.add(Dense(1, activation = 'sigmoid'))

model3.summary()
```

```python
model3.compile(loss = 'binary_crossentropy',
               optimizer = 'adam',
               metrics = ['accuracy'])
```

```python
history3 = model3.fit(x3_train, y3_train,
                      epochs = 150,
                      callbacks = [plot_loss],
                      verbose = 0,
                      validation_data = (x3_test, y3_test))
```

```python
score3 = model3.evaluate(x3_test, y3_test)
print('Test Loss:', round(score3[0], 2))
print('Test Accuracy:', round(score3[1], 2))
```
